```python
gender_columns_to_check = ["male", "female", "transgender", "other_gender"]
bias_df["gender_bias"] = (bias_df[gender_columns_to_check] > 0.5).any(axis=1).astype(int)

sexual_columns_to_check = ["heterosexual", "homosexual_gay_or_lesbian", "bisexual", "other_sexual_orientation"]
bias_df["sexual_bias"] = (bias_df[sexual_columns_to_check] > 0.5).any(axis=1).astype(int)

religion_columns_to_check = ["christian", "jewish", "muslim", "hindu", "buddhist", "atheist", "other_religion"]
bias_df["religion_bias"] = (bias_df[religion_columns_to_check] > 0.5).any(axis=1).astype(int)

race_columns_to_check = ["black", "white", "asian", "latino", "other_race_or_ethnicity"]
bias_df["race_bias"] = (bias_df[race_columns_to_check] > 0.5).any(axis=1).astype(int)

disability_columns_to_check = ["physical_disability", "intellectual_or_learning_disability", "psychiatric_or_mental_illness", "other_disability"]
bias_df["disability_bias"] = (bias_df[disability_columns_to_check] > 0.5).any(axis=1).astype(int)

# Make sure the toxicity annotation is accurate and not too high
filtered_bias_df = bias_df[bias_df['toxicity_annotator_count'] >= 10]
filtered_bias_df = filtered_bias_df[filtered_bias_df['toxicity'] < 0.25]

# Make sure the identity annotation is accurate
filtered_bias_df = filtered_bias_df[filtered_bias_df['identity_annotator_count'] >= 10]
filtered_bias_df = filtered_bias_df[['id', 'comment_text', 'gender_bias', 'sexual_bias', 'religion_bias', 'race_bias', 'disability_bias']]

print("Number or of rows after filters: " + str(len(filtered_bias_df)))

#filtered_bias_df = filtered_bias_df.sample(n=1000, random_state=42) 
filtered_bias_df = filtered_bias_df.dropna(subset=['comment_text'])

filtered_bias_df
```

```python
train_bias_df, test_bias_df = train_test_split(filtered_bias_df, test_size=0.3, random_state=42)
train_bias_df.to_csv('train_jigsaw_data.csv', index = False)
test_bias_df.to_csv('test_jigsaw_data.csv', index = False)
```
